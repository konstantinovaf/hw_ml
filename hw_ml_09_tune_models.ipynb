```python
# Choose algoritms which I'll use
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score


from sklearn import linear_model
from sklearn import svm
from sklearn import neighbors
from sklearn import tree

# I level algoritms
    # ridge
    # linearRegression
    # SVR
    # Nearest Neibours
    # Desicion tree regressor
    # SGD reg

# II level algoritms
    # voting regression
    # staked generalisation
    # RandomForest regression
    # xgboost

# 
simple_algorithms = {
    'ridge' : linear_model.Ridge(),
    'lin_reg' : linear_model.LinearRegression(),
    'svr' : svm.SVR(),
    'near_neib' : neighbors.KNeighborsRegressor(),
    'tree_reg' : tree.DecisionTreeRegressor(),
    'sgd_reg' : linear_model.SGDRegressor()
}

# ensemble_algorithms = {
#     linear_model.Ridge(alpha=.5),
#     linear_model.LinearRegression(),
#     'svr' : svm.SVR(),
#     'near_neib' : neighbors.KNeighborsRegressor(n_neighbors = 3, weights='uniform'),
#     'tree_reg' : tree.DecisionTreeRegressor(),
#     'sgd_reg' : linear_model.SGDRegressor(max_iter=1000, tol=1e-3)
# }
```

```python
# run I-level algoritms
for algoritm in simple_algorithms.values():
    model = algoritm.fit(X_train_scaled, y_train)
    model_name = type(model).__name__
    y_pred = algoritm.predict(X_test_scaled)
    MSE = mean_squared_error(y_test, y_pred)
    scoreCV = cross_val_score(algoritm, X_scaled, y, cv= 10, scoring='r2')
    score = algoritm.score(X_test_scaled, y_test)
    print(f'{model_name} -> MSE: {MSE}, score: {score}, scoreCV = {scoreCV.mean()}')

# Conclusion: 
# Models (Ridge, DecisionTreeRegressor, SGDRegressor) let's use for ensemble models    
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.neighbors import KNeighborsRegressor

estimators = [('ridge', linear_model.Ridge()),
    ('lin_reg', linear_model.LinearRegression()),
    ('svr', svm.SVR()),
    ('near_neib', neighbors.KNeighborsRegressor()),
    ('tree_reg', tree.DecisionTreeRegressor()),
    ('sgd_reg', linear_model.SGDRegressor())]

# create main model which includ 3 estimators
main_model = StackingRegressor(estimators=estimators, final_estimator=RandomForestRegressor(random_state=42))
main_model.fit(X_train_scaled, y_train)
y_pred_main = main_model.predict(X_test_scaled)
MSE = mean_squared_error(y_test, y_pred_main)
score = main_model.score(X_test_scaled, y_test)
scores = cross_val_score(main_model, X_train_scaled, y_train, cv=10)
print(f'MSE: {MSE}, score: {score}, scoreCV: {scores.mean()}, (+/- {scores.std() * 2}) ')

# Conclusion:
# ensemble models more sustainable and have better scoreCV
```

```python
# tune parameters
from sklearn.model_selection import GridSearchCV

models = [(linear_model.Ridge(), {'alpha': [0.5, 0.55, 0.6, 0.65, 0.7, 0.75], 
                                  'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'],
                                 'fit_intercept': ['True', 'False']}),
#           {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}),
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html
                                  
          (linear_model.LinearRegression(), {'fit_intercept': ['True', 'False'], 
                                            'normalize': ['True', 'False']}),
#  https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression         
          (svm.SVR(), {'gamma': [0.001, 0.01, 0.1, 1, 10, 100], 'C': [0.001, 0.01, 0.1, 1, 10, 100]}),
          
          (neighbors.KNeighborsRegressor(), {'n_neighbors':[1,2,3,4,5,10], 'weights':['uniform', 'distance']}),
          (tree.DecisionTreeRegressor(), {'max_depth': ['None', 5,10,15,20,25, 30, 40, 50, 100, 1000],
                                         'criterion': ['mse', 'friedman_mse', 'mae']}),
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
          
          (linear_model.SGDRegressor(), {'max_iter': [5, 10, 20,500,600, 700, 800, 900, 1000, 1100, 1200], 
                                         'penalty': ['l1','l2', 'elasticnet']})          ]

for alg, params in models:
    grid = GridSearchCV(alg, params, cv=5)
    grid.fit(X_train_scaled, y_train)
    print('Алгоритм: {}'.format(alg))
    print("Наилучшее значение перекрестной проверки: {:.2f}".format(grid.best_score_))
    print("Наилучшие параметры: ", grid.best_params_)
```
